```python
"""
Regine Benton
Module: 16.2
"""
import csv

# Reading the books file
with open('books.csv', mode='r', newline='') as file:
    reader = csv.DictReader(file)
    books = list(reader)
    
# Printing the book's values  
for book in books:
    print(book)


"""
Regine Benton
Module: 16.4
"""
import sqlite3

# Connecting to the SQLite Database
conn = sqlite3.connect('books.db')

# Inserting a record into the the book's table
cursor = conn.cursor()

# Running SQL commands against the database and Creating a table
cursor = conn.execute('''
CREATE TABLE IF NOT EXISTS books (
    title TEXT,
    author TEXT,
    year INTEGER
)
''')

# Committing changes and exiting connection
conn.commit()
conn.close()


"""
Module: 16.8
"""
from sqlalchemy import create_engine, Table, MetaData

# Connecting to the database
engine = create_engine('sqlite:///books.db')

# Creating an object to reflect the structure of the database
metadata = MetaData()

# Reflecting the 'books' table from the database
books_table = Table('books', metadata, autoload_with=engine)

# Connectng to the database and querying the titles in 'abc' order
with engine.connect() as connection:
    result = connection.execute(books_table.select().order_by(books_table.c.title))

    # Printing the titles
    for row in result:
        print(row['title'])
```
